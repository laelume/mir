```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt

hop_length = 1024
```

```python
f1="C:/Users/laelu/fma/data/fma_small/000/000002.mp3"
s1 , sr = librosa.load(f1,mono=True,sr=None)

f2="C:/Users/laelu/fma/data/fma_small/000/000005.mp3"
s2 , sr = librosa.load(f2,mono=True,sr=None)

f3="C:/Users/laelu/fma/data/fma_small/000/000010.mp3"
s3 , sr = librosa.load(f3,mono=True,sr=None)
```

```python
S1 = librosa.feature.melspectrogram(y=s1, sr=sr, n_mels=128)
log_S1 = librosa.power_to_db(S1, ref=np.max)

plt.figure(figsize=(12,4))
librosa.display.specshow(log_S1, sr=sr, x_axis='time', y_axis='mel')
plt.title('mps: RAP1')
plt.colorbar(format='%+02.0f dB')
plt.tight_layout()
```

```python
S2 = librosa.feature.melspectrogram(y=s2, sr=sr, n_mels=128)
log_S2 = librosa.power_to_db(S2, ref=np.max)

plt.figure(figsize=(12,4))
librosa.display.specshow(log_S2, sr=sr, x_axis='time', y_axis='mel')
plt.title('mps: RAP2')
plt.colorbar(format='%+02.0f dB')
plt.tight_layout()
```

```python
S3 = librosa.feature.melspectrogram(y=s3, sr=sr, n_mels=128)
log_S3 = librosa.power_to_db(S3, ref=np.max)

plt.figure(figsize=(12,4))
librosa.display.specshow(log_S3, sr=sr, x_axis='time', y_axis='mel')
plt.title('mps: ROCK1')
plt.colorbar(format='%+02.0f dB')
plt.tight_layout()
```

```python
sh1, sp1 = librosa.effects.hpss(s1)

S_H1   = librosa.feature.melspectrogram(sh1, sr=sr)
S_P1 = librosa.feature.melspectrogram(sp1, sr=sr)

log_SH1 = librosa.power_to_db(S_H1, ref=np.max)
log_SP1 = librosa.power_to_db(S_P1, ref=np.max)

plt.figure(figsize=(12,6))
plt.subplot(2,1,1)

librosa.display.specshow(log_SH1, sr=sr, y_axis='mel')
plt.title('mps Harmonic: RAP1')
plt.colorbar(format='%+02.0f dB')
plt.subplot(2,1,2)

librosa.display.specshow(log_SP1, sr=sr, x_axis='time', y_axis='mel')
plt.title('mps Percussive: RAP1')
plt.colorbar(format='%+02.0f dB')
plt.tight_layout()
```

```python
sh2, sp2 = librosa.effects.hpss(s2)

S_H2   = librosa.feature.melspectrogram(sh2, sr=sr)
S_P2 = librosa.feature.melspectrogram(sp2, sr=sr)

log_SH2 = librosa.power_to_db(S_H2, ref=np.max)
log_SP2 = librosa.power_to_db(S_P2, ref=np.max)

plt.figure(figsize=(12,6))
plt.subplot(2,1,1)

librosa.display.specshow(log_SH2, sr=sr, y_axis='mel')
plt.title('mps Harmonic: RAP2')
plt.colorbar(format='%+02.0f dB')
plt.subplot(2,1,2)

librosa.display.specshow(log_SP2, sr=sr, x_axis='time', y_axis='mel')
plt.title('mps Percussive: RAP2')
plt.colorbar(format='%+02.0f dB')
plt.tight_layout()
```

```python
sh3, sp3 = librosa.effects.hpss(s3)

S_H3   = librosa.feature.melspectrogram(sh3, sr=sr)
S_P3 = librosa.feature.melspectrogram(sp3, sr=sr)

log_SH3 = librosa.power_to_db(S_H3, ref=np.max)
log_SP3 = librosa.power_to_db(S_P3, ref=np.max)

plt.figure(figsize=(12,6))
plt.subplot(2,1,1)
librosa.display.specshow(log_SH3, sr=sr, y_axis='mel')
plt.title('mel power spectrogram: ROCK (Harmonic)')
plt.colorbar(format='%+02.0f dB')
plt.subplot(2,1,2)
librosa.display.specshow(log_SP3, sr=sr, x_axis='time', y_axis='mel')
plt.title('mel power spectrogram: ROCK (Percussive)')
plt.colorbar(format='%+02.0f dB')
plt.tight_layout()
```

```python
mfcc1 = librosa.feature.mfcc(y=s1, sr=sr, hop_length=hop_length, n_mfcc=13)
librosa.display.specshow(mfcc1[1:12,:],x_axis='time')
plt.ylabel('MFCC s1: RAP (1:256)')
```

```python
mfcc01        = librosa.feature.mfcc(S=log_S1, n_mfcc=13)
delta_mfcc01  = librosa.feature.delta(mfcc1[1:12])
delta2_mfcc01 = librosa.feature.delta(mfcc1[1:12], order=2)

plt.figure(figsize=(12, 6))
plt.subplot(3,1,1)
librosa.display.specshow(mfcc1[1:12])
plt.ylabel('MFCC1: RAP (2:13)')
plt.colorbar()
plt.subplot(3,1,2)
librosa.display.specshow(delta_mfcc01)
plt.ylabel('MFCC1-$\Delta$')
plt.colorbar()
plt.subplot(3,1,3)
librosa.display.specshow(delta2_mfcc01, sr=sr, x_axis='time')
plt.ylabel('MFCC1-$\Delta^2$')
plt.colorbar()
plt.tight_layout()
```
